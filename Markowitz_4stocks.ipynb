```python
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
from pandas_datareader import data as wb
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
CPALL = wb.DataReader('CPALL.BK', data_source = 'yahoo', start='2017-1-1')
```

```python
CBG = wb.DataReader('CBG.BK', data_source = 'yahoo', start= '2017-1-1')
```

```python
DTAC = wb.DataReader('DTAC.BK', data_source='yahoo', start='2017-1-1')
```

```python
EGCO = wb.DataReader('EGCO.BK', data_source='yahoo', start='2017-1-1')
```

```python
EGCO
```

```python
data = pd.concat([CPALL.iloc[:,3], CBG.iloc[:,3], DTAC.iloc[:,3], EGCO.iloc[:,3]], axis=1)
```

```python
data
```

```python
data.columns = ['CPALL', 'CBG', 'DTAC', 'EGCO']
```

```python
data
```

```python
((data/data.iloc[0])*100).plot(figsize = (10,6))
```

```python
log_returns = np.log(1+ data.pct_change(1))
log_returns
```

```python
log_returns.plot(figsize=(10,6))
```

```python
returns = log_returns.mean()*250
returns
```

```python
cov = log_returns.cov()*250
cov
```

```python
corr = log_returns.corr()
corr
```

```python
log_returns.describe()
```

```python
num_assets = 4
weight0 = []
weight1 = []
weight2 = []
weight3 = []
port_returns = []
port_volatility = []
```

```python
for i in range(10000):
    weight = np.random.random(num_assets)
    weight /= np.sum(weight)
    returns = np.sum(weight*log_returns.mean())*250
    volatility = np.sqrt(np.dot(np.dot(log_returns.cov()*250,weight),weight.T))
    weight_0 = weight[0]
    weight_1 = weight[1]
    weight_2 = weight[2]
    weight_3 = weight[3]
    port_returns.append(returns)
    port_volatility.append(volatility)
    weight0.append(weight_0)
    weight1.append(weight_1)
    weight2.append(weight_2)
    weight3.append(weight_3)
```

```python
port_returns = np.array(port_returns)
port_volatility = np.array(port_volatility)
weight0 = np.array(weight0)
weight1 = np.array(weight1)
weight2 = np.array(weight2)
weight3 = np.array(weight3)
```

```python
portfolios = pd.DataFrame({'Returns': port_returns, 'Volatility':port_volatility, 'CPALL':weight0, 'CBG':weight1, 'DTAC':weight2, 'EGCO':weight3})
```

```python
portfolios
```

```python
portfolios.plot(x='Volatility', y='Returns', kind ='scatter', figsize=(12,8))
```
